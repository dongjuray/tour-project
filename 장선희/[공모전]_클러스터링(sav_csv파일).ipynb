## 국가별 집계

```python
d2024['D_NAT'] = d2024['D_NAT'].str.replace(' ', '', regex=False)
```

```python
import pandas as pd

# 1. 데이터 불러오기
df = d2024.copy()
dtype_df = dtype.copy()

# 2. 변수 분류
binary_vars = dtype_df[dtype_df["수정된 분류"] == "이진"]["변수명"].tolist()
categorical_vars = dtype_df[dtype_df["수정된 분류"] == "범주형"]["변수명"].tolist()
numeric_vars = dtype_df[dtype_df["수정된 분류"] == "수치형"]["변수명"].tolist()

# 3. 국가 구분 변수 지정 및 결측 제거
nationality_col = "D_NAT"
df_national = df.dropna(subset=[nationality_col])

# 4. 실제 존재하는 변수만 필터링
binary_vars = [col for col in binary_vars if col in df_national.columns]
categorical_vars = [col for col in categorical_vars if col in df_national.columns]
numeric_vars = [col for col in numeric_vars if col in df_national.columns]

# 5. 범주형 변수 → One-hot 인코딩
categorical_encoded = pd.get_dummies(df_national[categorical_vars])
categorical_encoded[nationality_col] = df_national[nationality_col]

# 6. 변수 유형별 국가별 평균 계산
binary_df = df_national.groupby(nationality_col)[binary_vars].mean(numeric_only=True)
categorical_df = categorical_encoded.groupby(nationality_col).mean(numeric_only=True)
numeric_df = df_national.groupby(nationality_col)[numeric_vars].mean(numeric_only=True)

# 7. 최종 통합
national_pca_df = pd.concat([binary_df, categorical_df, numeric_df], axis=1)

# 8. 저장 (선택)
national_pca_df.to_csv("국가별_PCA_입력데이터.csv")
```

## PCA 전처리

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


# 2. 결측값 평균으로 대체
national_pca_df_clean = national_pca_df.fillna(national_pca_df.mean())

# 3. 표준화 (스케일링)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(national_pca_df_clean)

# 4. PCA 수행 (누적 설명력 95% 기준)
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)

# 5. 각 변수의 주성분 기여도 (로딩값)
loadings = pd.DataFrame(
    pca.components_.T,  # 행: 변수, 열: PC들
    index=national_pca_df_clean.columns,
    columns=[f"PC{i+1}" for i in range(pca.n_components_)]
)

# 6. 각 국가의 주성분 좌표값
national_pca_coords = pd.DataFrame(
    X_pca,
    index=national_pca_df.index,  # 국가명
    columns=[f"PC{i+1}" for i in range(pca.n_components_)]
)
```

### LOADINGS

각 원 변수들이 주성분(PC1, PC2, ...)에 기여하는 정도를 나타냄

🔍 해석 방법
PC1: 체류일수, 쇼핑지출과 관련 → 체류 활동 성향

PC2: 음식체험 중심 → 체험 위주 성향

부호와 크기 모두 중요: +0.8 vs -0.8은 방향만 다를 뿐 중요도는 동일

🧠 활용 예시
주성분의 의미를 해석해 이름 붙이기:
예: PC1 → "소비 중심형", PC2 → "체험 중심형"

### national_pca_coords

각 국가가 주성분 공간에서 어디에 위치하는지를 나타냄

🔍 해석 방법
PC1~3의 위치로 국가 간의 특성 차이를 비교

예:

PC1 점수가 높을수록 → 장기체류, 고소비형

PC2 점수가 높을수록 → 체험/문화중심 여행 성향

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 25))

# PC1~PC5 각각에 대해 subplot 생성
for i, pc in enumerate(loadings.columns[:5], start=1):
    # 상위 10개 변수 추출 (절댓값 기준)
    top_vars = loadings[pc].abs().sort_values(ascending=False).head(10).index
    top_vals = loadings.loc[top_vars, pc].sort_values()  # 기여도 방향 유지

    plt.subplot(5, 1, i)
    plt.barh(top_vals.index, top_vals.values, color='mediumseagreen')
    plt.title(f"{pc}의 주요 기여 변수 Top10")
    plt.xlabel("로딩값")
    plt.ylabel("변수명")

plt.tight_layout()
plt.show()
```

```python
# 각 PC별 상위 10개 변수 추출
top_n = 10
top_vars_per_pc = {}

for pc in loadings.columns[:5]:  # PC1 ~ PC5
    top_vars = loadings[pc].abs().sort_values(ascending=False).head(top_n)
    top_vars_per_pc[pc] = loadings.loc[top_vars.index, pc]

# 보기 쉽게 출력
for pc, top_vars in top_vars_per_pc.items():
    print(f"\n[{pc} 상위 {top_n} 변수]")
    print(top_vars)
```

## 클러스터링

```python

# 1. 사용할 컬럼
plot_cols = ["PC1", "PC2", "PC3", "PC4", "PC5"]
n = len(plot_cols)

# 2. 전체 subplot 그리기 (대각선 기준 아래쪽만, corner=True 효과)
fig, axes = plt.subplots(n, n, figsize=(15, 15))
fig.suptitle("국가별 PCA 좌표 분포 (PC1~PC5)", y=0.92)

for i in range(n):
    for j in range(n):
        ax = axes[i, j]

        if i < j:
            ax.axis('off')  # upper triangle은 제거
        elif i == j:
            ax.hist(coords[plot_cols[i]], bins=15, color='skyblue')
            ax.set_ylabel('')
            ax.set_xlabel(plot_cols[i])
        else:
            ax.scatter(coords[plot_cols[j]], coords[plot_cols[i]], s=20, alpha=0.7, color='gray')
            if j == 0:
                ax.set_ylabel(plot_cols[i])
            else:
                ax.set_ylabel('')
            if i == n - 1:
                ax.set_xlabel(plot_cols[j])
            else:
                ax.set_xlabel('')

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
# 1. PCA 좌표 불러오기
coords

# 2. 분석 범위 설정
p_range = range(2, 11)  # 주성분 수: 2~10
k_range = range(2, 9)   # 클러스터 수: 2~8

# 3. 결과 저장용 데이터프레임 생성
silhouette_scores = pd.DataFrame(index=p_range, columns=k_range)
wcss_scores = pd.DataFrame(index=p_range, columns=k_range)

# 4. P x K 조합별로 점수 계산
for p in p_range:
    X = coords.iloc[:, :p].values
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = kmeans.fit_predict(X)

        # Silhouette score
        sil_score = silhouette_score(X, labels)
        silhouette_scores.loc[p, k] = sil_score

        # WCSS
        wcss_scores.loc[p, k] = kmeans.inertia_

# 실루엣 점수 히트맵
plt.figure(figsize=(10, 6))
data = silhouette_scores.astype(float).values

im = plt.imshow(data, cmap='YlGnBu', aspect='auto')

# x, y축 라벨
plt.xticks(ticks=np.arange(len(k_range)), labels=list(k_range))
plt.yticks(ticks=np.arange(len(p_range)), labels=list(p_range))
plt.xlabel("군집 수 (K)")
plt.ylabel("주성분 수 (P)")
plt.title("Silhouette Score Heatmap\n(P: 주성분 수, K: 군집 수)")

# 값 표시
for i in range(len(p_range)):
    for j in range(len(k_range)):
        val = data[i, j]
        plt.text(j, i, f"{val:.3f}", ha='center', va='center', color='black')

plt.colorbar(im, label='Silhouette Score')
plt.tight_layout()
plt.show()

# Elbow Curve
plt.figure(figsize=(10, 6))
for p in p_range:
    y_vals = wcss_scores.loc[p].astype(float).values
    plt.plot(k_range, y_vals, marker='o', label=f"P={p}")

plt.title("Elbow Curve (WCSS vs. K) by PCA dimension")
plt.xlabel("클러스터 수 (K)")
plt.ylabel("WCSS (총 제곱거리)")
plt.grid(True)
plt.legend(title="주성분 수 (P)")
plt.tight_layout()
plt.show()
```
